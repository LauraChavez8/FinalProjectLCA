```python
finaldf.customer_status_final = finaldf.customer_status_final.replace('current','0')
finaldf.customer_status_final = finaldf.customer_status_final.replace('pre-loss','1')
finaldf.customer_status_final = finaldf.customer_status_final.replace('loss','2')
```

```python
finaldf.info()
```

```python
finaldf.customer_status_final = finaldf.customer_status_final.astype('int64')
```

```python
df_prueba_pivot = finaldf.pivot_table(index= 'account__id', columns=['date'], values=['days_late_at_date'],
                                       aggfunc="max")

df_prueba_pivot.head(10)
```

```python
df_prueba_pivot = df_prueba_pivot.droplevel(0, axis=1)
```

```python
extracol = finaldf[['account__id','customer_status_final']]

extracol.head(10)
```

```python
extracol.shape
```

```python
extracol.isna().sum()
```

```python
custcol = extracol.groupby(by='account__id').max()

custcol.head(10)
```

```python
custcol.value_counts()
```

```python
finaldata = custcol.merge(df_prueba_pivot, on='account__id')

finaldata.head(10)
```

```python
finaldata.isna().sum()
```

```python
finaldata = finaldata.fillna(0)
```
