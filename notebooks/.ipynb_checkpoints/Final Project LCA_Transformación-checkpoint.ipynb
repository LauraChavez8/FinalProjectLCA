```python
pruebadf.customer_status_final = pruebadf.customer_status_final.replace('current','0')
pruebadf.customer_status_final = pruebadf.customer_status_final.replace('pre-loss','1')
pruebadf.customer_status_final = pruebadf.customer_status_final.replace('loss','2')
```

```python
pruebadf.info()
```

```python
pruebadf.customer_status_final = pruebadf.customer_status_final.astype('int64')
```

```python
df_prueba_pivot = pruebadf.pivot_table(index= 'account__id', columns=['date'], values=['days_late_at_date'],
                                       aggfunc="max")

df_prueba_pivot
```

```python
df_prueba_pivot = df_prueba_pivot.droplevel(0, axis=1)
```

```python
extracol = pruebadf[['account__id','customer_status_final']]

extracol
```
